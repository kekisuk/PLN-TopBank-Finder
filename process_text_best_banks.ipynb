```python
import pandas as pd
import json
from unidecode import unidecode
from spellchecker import SpellChecker
import ast
import re
from collections import Counter
```

Lendo arquivo json

```python
with open("discussion_best_banks.json", "r", encoding="utf-8") as f:
    data = json.load(f)

structured_data = []
```

```python
for idx, post in enumerate(data, start=1):
    titulo = post["title"]
    comentarios = post.get("comments", [])
    qtd_comentarios = len(comentarios)

    structured_data.append({
        "id": idx,
        "title": titulo,
        "qtd_comentarios": qtd_comentarios,
        "comentarios": json.dumps(comentarios, ensure_ascii=False)  # Mantendo em formato JSON
    })
```

```python
df = pd.DataFrame(structured_data)
display(df)
```

Processamento e Limpeza

```python
# Inicializando o corretor ortográfico
spell = SpellChecker(language="pt")
```

```python
# Função para limpar e corrigir texto
def clean_text(text):
    text = unidecode(text)  # Remover acentos
    text = re.sub(r'[^\w\s]', '', text)  # Remover emojis e caracteres especiais
    words = text.split()
    corrected_words = [spell.correction(word) if spell.correction(word) else word for word in words]
    return " ".join(corrected_words)
```

```python
# Função para processar lista de comentários
def clean_comments(comment_list):
    try:
        comments = ast.literal_eval(comment_list)  # Converter string para lista
        return [clean_text(comment) for comment in comments]
    except (SyntaxError, ValueError):
        return []
```

```python
# Função para normalizar o texto (remover acentuação e converter para minúsculas)
def normalize_text(text):
    text = unidecode(text)  # Remover acentos
    text = text.lower()  # Converter para minúsculas
    return text
```

```python
# Aplicando a limpeza
df["title"] = df["title"].apply(clean_text)
```

Contando frequência dos bancos

```python
# Carregando a lista de bancos a partir do CSV
df_banks = pd.read_csv('archives/banks_list.csv')

# Convertendo a coluna 'Bank' para uma lista
banks = df_banks['Bancos'].tolist()
```

```python
banks
```

```python
# Função para contar a frequência dos bancos nos comentários (ignorando acentuação e maiúsculas/minúsculas)
def count_banks(comments):
    bank_counter = Counter()
    for comment in comments:
        for bank in banks:
            normalized_bank = normalize_text(bank)  # Normalizar o nome do banco
            if normalized_bank in comment:  # Verificar se o banco está no comentário
                bank_counter[bank] += 1
    return dict(bank_counter)
```

```python
# Criando a nova coluna 'banks_list' que contém a contagem dos bancos nos comentários
df["banks_list"] = df["comentarios"].apply(count_banks)
```
